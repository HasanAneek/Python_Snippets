```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
data = pd.read_csv('data3.csv')
view_count = data['view_count'] 
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolors='black', linewidths=0.75)
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio!')
plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')


plt.tight_layout()










# SAMPLE DATA
# x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
# y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
# colors = [1, 2, 4, 8, 3, 5, 9, 2, 3, 4, 5, 6, 7, 7, 8, 3, 3, 2, 3, 4]
# sizes = [123, 567, 763, 237, 834, 671, 763, 986, 459, 126, 158, 747, 467, 389, 578, 637, 
#          997,775,336,446]

# # marker='X'
# plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', alpha=0.75, linewidths=0.75, 
#             edgecolors='black')  
# cbar = plt.colorbar()
# cbar.set_label('Satisfaction')
```
